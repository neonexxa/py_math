```python
import requests
from bs4 import BeautifulSoup
import re
# from urllib import request
import os
from six.moves import urllib

# film-showtimes
main_url = "https://tickets.mmcineplexes.com/Browsing/Cinemas/Details/1000"
page = requests.get(main_url)
soup = BeautifulSoup(page.content, 'html.parser')
filems = soup.find_all(class_="film-showtimes")
eachfilemarray = [filem for filem in filems]
for filem in filems:
#     loopeach movie
    print ("filem title: " ,filem.find(class_="film-title").getText())
    for showday in filem.find_all(class_="session"):
        print("session date:",showday.find(class_="session-date").getText())
        for idx,showtime in enumerate(showday.find_all(class_="session-time")):
            print("showtime ",idx+1," ",showtime.find('time').getText())
```
