```python
import tkinter as tk
import tkinter.messagebox
from tkinter import *
import requests
from bs4 import BeautifulSoup
from PIL import Image, ImageTk
import io
import base64
from urllib.request import HTTPError
from urllib.request import urlopen
from io import StringIO,BytesIO
import urllib

class SampleApp(tk.Tk):
    def __init__(self):
        tk.Tk.__init__(self)
        self.entry = tk.Entry(self)
        self.button = tk.Button(self, text="Get", command=self.on_button)
        self.button.pack()
        self.entry.pack()



    def on_button(self):
        url = "http://erolord.com/?s="+self.entry.get()
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')
        mason = soup.find(class_='masonry').find_all(class_='thumb')
        arr = [child.find(class_='aa1')['href'] for child in mason if child.find(class_='aa1')]
        arr1 = [child.find(class_='aa1') for child in mason if child.find(class_='aa1')]
        hentais = [hentais.findChildren() for hentais in arr1]
        hentai_pics = [child[0]['src'] for child in hentais]
        print(hentai_pics)
#         canvas = tk.Canvas(self,width=999,height=999)
#         canvas.pack()
#         cv.pack(side='top', fill='both', expand='yes')
        photos = []
        for hentai in hentai_pics:
            image = Image.open(io.BytesIO(urlopen(hentai).read())).resize((200,200),Image.ANTIALIAS) #.resize(10, 10)
#             photo = tk.PhotoImage(image)
            photos.append(image)
        new_winF(self,photos)

def load_image_to_base64(image_url):
    """ Load an image from a web url and return its data base64 encoded"""
    image_byt = urlopen(image_url).read()
    image_b64 = base64.encodestring(image_byt)
    return image_b64
def new_winF(self,arr): # new window definition
    newwin = Toplevel(self)
    display = Label(newwin, text="Humm, see a new window !")
    display.pack()  
    img = {}
    ygrid = 0
    xgrid = 0
    for idx,gmba in enumerate(arr):
        print(idx,gmba)
        render = ImageTk.PhotoImage(gmba) #.subsample(4, 4)
        img[idx] = tk.Label(newwin, image=render)
#         img[idx].grid(row=idx)
        img[idx].image = render 

        xgrid = 0 if idx % 6 == 0 else xgrid
        xplace = 10+200*xgrid
        yplace = 10+200*ygrid+20*ygrid
        print(xplace,yplace)
        img[idx].place(x=xplace, y=yplace)
        ygrid = ygrid+1 if idx % 6 == 0 else ygrid
        xgrid+=1
        print(idx)
#         display.im1 = Image.open(arr[0])
    
app = SampleApp()
app.mainloop()
```
